## Feature augmentation : Lag features


```python
descriptive_list = ['force_x_mean', 'force_x_std',
            'force_y_mean', 'force_y_std',
            'force_z_mean', 'force_z_std',
            'acc_x_mean', 'acc_x_std',
            'acc_y_mean', 'acc_y_std',
            'acc_z_mean', 'acc_z_std',
            'ae_mean', 'ae_std']

df1 = df.iloc[:315,:]
df2 = df.iloc[315:630,:]

for col in descriptive_list:
    df1[f'{col}_lag-1'] = df1[col].shift(1)
    df1[f'{col}_lag-1'].fillna(value=0, inplace=True)
    df2[f'{col}_lag-1'] = df2[col].shift(1)
    df2[f'{col}_lag-1'].fillna(value=0, inplace=True)
    df_test[f'{col}_lag-1'] = df_test[col].shift(1)
    df_test[f'{col}_lag-1'].fillna(value=0, inplace=True)

df = pd.concat([df1, df2], axis=0)
```

```python
def plot_pearson_corr(df):

  '''Cria a matriz de correlaÃ§Ã£o entre certas features e as vÃ¡riÃ¡veis de interesse

  Parameters
  ----------
  df -> dataframe

  Output
  ------
  matriz
  '''
  cmap = sns.diverging_palette(h_neg=10,h_pos=240, as_cmap=True)
  corr = df.corr()
  mask = np.triu(np.ones_like(corr, dtype=bool))
  plt.figure(figsize=(12,8))
  sns.heatmap(df.corr(),
              mask=mask,center=0,
              cmap=cmap,
              linewidths=1,
              annot=True,
              fmt=".2f")
  plt.show()
```
